# 📝 Array Exercises

This notebook contains practice exercises for NumPy arrays.  
Try solving each exercise on your own before checking solutions (solutions are not provided here, only questions).

---

## 1. Array Creation

1. Create a 1D NumPy array of integers from 0 to 9.
2. Create a 2D array of shape (3, 3) filled with all `True`.
3. Generate an array of 10 random integers between 5 and 50.
4. Create an identity matrix of size 4×4.
5. Generate an array of 50 evenly spaced numbers between 0 and 5.

---

## 2. Array Attributes & Indexing

1. Given an array `arr = np.arange(1, 21).reshape(4, 5)`:
   - Find its shape, dimensions, size, and dtype.
   - Extract the 2nd row.
   - Extract the last column.
   - Slice the subarray containing rows 2–3 and columns 2–4.
2. Reverse the array `arr = np.arange(10)` using slicing.

---

## 3. Array Operations

1. Given two arrays `a = np.array([1, 2, 3])` and `b = np.array([4, 5, 6])`:
   - Perform element-wise addition, subtraction, multiplication, and division.
   - Compute the dot product.
2. Compute the sum, mean, and standard deviation of `np.arange(1, 11)`.
3. Square every element in an array without using a loop.

---

## 4. Array Manipulation

1. Reshape the array `np.arange(12)` into a 3×4 matrix.
2. Flatten a 2D array into a 1D array.
3. Horizontally stack `a = np.array([1, 2, 3])` and `b = np.array([4, 5, 6])`.
4. Split the array `np.arange(10)` into 2 equal parts.

---

## 5. Broadcasting

1. Add a 1D array `[1, 2, 3]` to each row of a 3×3 matrix of ones.
2. Multiply a column vector `[[1], [2], [3]]` with a row vector `[1, 2, 3]`.

---

## 6. Challenge Problems

1. Create a 5×5 array with values 1 to 25, then extract its diagonal elements.
2. Normalize an array so that all values lie between 0 and 1.
3. Replace all odd numbers in `np.arange(10)` with -1.
4. Generate a checkerboard pattern (8×8) using NumPy.
5. Create a 2D array where the border elements are 1 and the inside elements are 0.
