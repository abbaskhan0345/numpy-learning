## Step 1: Generate Synthetic Data

Weâ€™ll simulate student grades for multiple subjects using NumPyâ€™s random module.


```python
import numpy as np

# Reproducibility
np.random.seed(42)

# Parameters
num_students = 100
subjects = ["Math", "Physics", "Chemistry", "Biology", "English"]

# Generate random grades between 50 and 100
grades = np.random.randint(50, 101, size=(num_students, len(subjects)))

print("Shape of grades data:", grades.shape)
grades[:5]  # preview first 5 students
```


---



## Step 2: Compute Basic Statistics

Weâ€™ll calculate average, min, max, and standard deviation for each subject.


```python
# Mean per subject
mean_grades = grades.mean(axis=0)

# Min, Max, Std
min_grades = grades.min(axis=0)
max_grades = grades.max(axis=0)
std_grades = grades.std(axis=0)

# Display results
for i, subj in enumerate(subjects):
    print(f"{subj}: mean={mean_grades[i]:.2f}, min={min_grades[i]}, max={max_grades[i]}, std={std_grades[i]:.2f}")
```

---

## Step 3: Identify Top Performers

Weâ€™ll find the students with the highest total scores.

```python

# Total score for each student
total_scores = grades.sum(axis=1)

# Index of top 5 students
top_indices = total_scores.argsort()[-5:][::-1]

print("Top 5 students (index and total score):")
for idx in top_indices:
    print(f"Student {idx}: Total = {total_scores[idx]}")
```

---


## Step 4: Correlation Between Subjects (Linear Algebra)

Weâ€™ll check how strongly subjects are correlated.


```python
# Compute correlation matrix
corr_matrix = np.corrcoef(grades.T)

print("Correlation Matrix:")
print(corr_matrix)
```
